```python
import matplotlib.pyplot as plt
x=[20,30,40,50]
mylabels=["Apples","Bananas","Cherries","Dates"]
mycolors=["black","pink","blue","green"]
myexplode=[0.3,0,0,0]
plt.pie(x,labels=mylabels,colors=mycolors,explode=myexplode)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
mycolor=np.array(["r","b","y","g"])
plt.barh(x,y,color=mycolor)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
mycolor=np.array(["r","b","y","g"])
sizes=[100,200,150,300]
plt.title("scatter",fontsize=20)
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.scatter(x,y,c=mycolor,s=sizes,alpha=0.5,marker="*",edgecolor="black")
plt.show()
```

```python
plt.subplot(2,2,1)
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.subplot(2,2,2)
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.scatter(x,y)
plt.subplot(2,2,3)
x=np.array(["A","B","C","D"])
plt.pie([3,8,1,10])
plt.subplot(2,2,4)
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.plot(x,y)
plt.show()
```
