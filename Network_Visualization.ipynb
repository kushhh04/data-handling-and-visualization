```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.DiGraph()
G.add_nodes_from([1,2,3,4,5])
G.add_edges_from([(1,2),(1,3),(2,3),(3,4),(5,1)])
pos=nx.spring_layout(G)
nx.draw(G,pos,with_labels=True,node_color='black',node_size=1400,edge_color='red',linewidths=1,font_size=15,font_color='red')
plt.show()
```
